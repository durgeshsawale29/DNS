```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('heart.csv')
```

```python
df1
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.barplot(data=df1,x='sex',y='chol')
plt.xlabel('sex')
plt.ylabel('chol')
plt.title('graph')
plt.show
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.scatterplot(data=df1,x='age',y='chol')
plt.xlabel('age')
plt.ylabel('chol')
plt.title('graph')
plt.show
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.histplot(data=df1,x='age')
plt.xlabel('age')
plt.title('graph')
plt.show
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.boxplot(data=df1,x='cp',y='chol')
plt.xlabel('cp')
plt.ylabel('chol')
plt.title('graph')
plt.show
```
