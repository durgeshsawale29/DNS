```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset_facebook.csv',sep=';')
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.barplot(data=df,x='Type',y='Total Interactions')
plt.xlabel('Type')
plt.ylabel('total interactions')
plt.title('graph')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
sns.scatterplot(data=df,x='Lifetime Post Total Reach',y='Lifetime Post Total Impressions',alpha=1)
plt.xlabel('Type')
plt.ylabel('total interactions')
plt.title('graph')
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.subplot(2,2,2)
sns.histplot(data=df,x='Lifetime Engaged Users',bins=500,kde=False,color='pink')
plt.xlabel('Type')
plt.ylabel('total interactions')
plt.title('graph')
plt.show()
```

```python

plt.subplot(2,2,2)
sns.boxplot(data=df,x='Category',y='Lifetime Post Consumers')
plt.xlabel('Type')
plt.ylabel('total interactions')
plt.title('graph')
plt.show()
```
